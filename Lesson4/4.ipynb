```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

4. Решите аналитически и потом численно (в программе) уравнение, зависящее от параметра а: 
sin(а\*x)=0
при условии: 0.01<a<0.02, 100<х<500.
Т.е. надо найти решение х как функцию параметра а - построить график x=x(а).
Если численным методом не получается найти все ветви решения x(а), то отыщите хотя бы одну.

```python
a = np.linspace(0.01, 0.02, 100)
x1 = np.pi/a
x2 = 2*np.pi/a
x3 = 3*np.pi/a
```

```python
plt.figure(figsize=(8,5))
plt.plot(a,x1,label='n=1')
plt.plot(a,x2,label='n=2')
plt.plot(a,x3,label='n=3')
plt.axhline(100, c='k', ls='--')
plt.axhline(500, c='k', ls='--')
plt.legend()
```

```python
def f1(a):
    return 2*np.pi/a - 500

def f2(a):
    return 3*np.pi/a - 500

plt.figure(figsize=(8,5))
plt.plot(a,x1,label='n=1')
plt.plot(a,x2,label='n=2')
plt.plot(a,x3,label='n=3')
plt.axhline(100, c='r')
plt.axhline(500, c='r')
plt.axvline(fsolve(f1, 0.012), ymax=0.5, c='k', ls='--')
plt.axvline(fsolve(f2, 0.018), ymax=0.5, c='k', ls='--')
plt.xticks([0.01, round(fsolve(f1, 0.012)[0], 4), round(fsolve(f2, 0.018)[0], 4), 0.02])
plt.yticks([100, np.pi/0.02, np.pi/0.01, 3*np.pi/0.02, 500])
plt.ylim(0,1000)
plt.legend()
```

Ответ:
* $a \in (0.01; 0.02), x \in (157.1; 314.2)$
* $a \in (0.0126; 0.02), x \in (314.2; 500)$
* $a \in (0.0188; 0.02), x \in (471.2; 500)$
