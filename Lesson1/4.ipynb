```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
def func(k,x):
    return np.cos(k*x)

x = np.linspace(-10, 10, 200)
y1 = func(k=1, x=x)
y2 = func(k=4, x=x)
```

```python
plt.figure(figsize=(10,6))

plt.ylim(-2,2)
plt.xlim(-10,10)

plt.xlabel('x', {'size':18})
plt.ylabel('y(k,x)', {'size':18})

plt.plot(x, y1, label='k=1')
plt.plot(x, y2, label='k=4')
plt.axhline(0, c='k')
plt.axvline(0, c='k')

plt.legend(loc=1)
```
