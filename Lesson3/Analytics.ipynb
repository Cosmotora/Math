1.2 Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def sum_vectors(a, b):
    assert all([type(var) is list for var in [a,b]]), 'vectors must be list type'
    assert len(a) == len(b), 'vectors must be same size'
    return [i+j for i, j in zip(a,b)]
```

```python
a = [10, 10, 10]
b = [0, 0, -10]
sum_vectors(a, b)
```

3. Напишите код на Python, реализующий построение графиков:
* окружности,	
* эллипса,
* гиперболы.

```python
x0 = np.linspace(-1,1,2000)
x0 = np.append(x0, np.nan)
```

```python
a = 1.5
b = 1
r = 1
x = x0 + a

y1 = list(np.sqrt(r**2 - (x-a)**2) + b)
y2 = list(-np.sqrt(r**2 - (x-a)**2) + b)

plt.figure(figsize=(5,5))
plt.plot(list(x)*2, y1 + y2)
plt.plot(a,b,'r.', ms=10)

plt.axhline(0, c='k', lw=1)
plt.axvline(0, c='k', lw=1)
plt.xlim(-3,3)
plt.ylim(-3,3)
```

```python
a = 1.5
b = 1
c = np.sqrt(a**2-b**2)
x = x0 * a

y1 = list(np.sqrt((1 - (x/a)**2) * b**2))
y2 = list(-np.sqrt((1 - (x/a)**2) * b**2))

plt.figure(figsize=(5,5))
plt.plot(list(x)*2, y1 + y2)
plt.plot([-c, c],[0, 0],'r.', ms=10)

plt.axhline(0, c='k', lw=1)
plt.axvline(0, c='k', lw=1)
plt.xlim(-3,3)
plt.ylim(-3,3)
```

```python
a = 1.5
b = 1
c = np.sqrt(a**2+b**2)
x = x0 * a * 2

y1 = list(np.sqrt(((x/a)**2 - 1) * b**2))
y2 = list(-np.sqrt(((x/a)**2 - 1) * b**2))

plt.figure(figsize=(5,5))
plt.plot(list(x)*2, y1 + y2)
plt.plot([-c, c],[0, 0],'r.', ms=10)

plt.axhline(0, c='k', lw=1)
plt.axvline(0, c='k', lw=1)
plt.xlim(-3,3)
plt.ylim(-3,3)
```

5.1 Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
X, Y = np.meshgrid(X, Y)
Z2 = 4*X + 2*Y - 3
Z1 = 4*(X+8) + 2*(Y+15)
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
plt.show()
```

5.2 Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
X, Y = np.meshgrid(X, Y)
Z = 4*(X**2) - 8*(Y**2) + 15
ax.plot_surface(X, Y, Z)
plt.show()
```
