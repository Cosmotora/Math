1. Нарисуйте график функции y = k * np.cos(x - a) + b  для некоторых (2-3 различных) значений параметров k, a, b

```python
def my_func(k, a, b, x):
    y = k * np.cos(x - a) + b
    return y

x = np.linspace(-5,5,500)
```

4.1 Решите систему уравнений:<br>
exp(x) + x∙(1 – y) = 1<br>
y = x2 – 1

```python
from scipy.optimize import fsolve

x = np.linspace(-4, 5, 201)
```

Произведем замену $y$ в первом уравнении и найдем его корни:

```python
def f1(x):
    return np.exp(x) + x * (1 - x**2 + 1) - 1

def f2(x):
    return x**2 - 1

plt.plot(x, f1(x))
plt.axhline(0, c='k', lw=1)

x_roots = []
y_roots = []

for i, idx in enumerate([-2, 0, 2, 4], 1):
    x_ = fsolve(f1, idx)[0]
    y_ = f2(x_)
    plt.plot(x_, f1(x_), 'r.')
    plt.text(x_-0.1, f1(x_), i, ha='right', va='bottom')
    print(f'x{i} = {x_:.4f}, y{i} = {y_:.4f}')
    x_roots.append(x_)
    y_roots.append(y_)
```

4.2 Решите систему уравнений и неравенств:<br>
y = x2 – 1<br>
exp(x) + x∙(1 – y)  - 1 > 0

Подставим $y$ из первого уравнения в неравенство, и обратимся к графику из предыдущего задания. Нас будут интересовать области $(-\infty; x_1) \cup (x_2; x_3) \cup (x_4; +\infty)$. Для самопроверки построим график $y = x^2 - 1$, чтобы не запутаться с областями для $y$.

```python
plt.plot(x, f2(x))
plt.xticks(x_roots)
plt.yticks(y_roots)

plt.grid()
```

Итого, корнями системы являются:
* $x \in (-\infty; -1.5818), y \in (1.5022; +\infty)$
* $x \in (0; 2.6181), y \in (-1; 5.8547)$
* $x \in (4.2001; +\infty), y \in (16.6408; +\infty)$
